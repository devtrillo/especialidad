```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12)

x = np.linspace(0, 1, 20)
y = np.cos(x) + 0.3*np.random.rand(20)
p = np.poly1d(np.polyfit(x, y, 10))

t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("./tabla-covid-cisneros.csv",
                 skiprows=[i for i in range(1, 2)])
df = df['TotalCases']
y = df.dropna()

x = np.array(range(0, 53))
p = np.poly1d(np.polyfit(x, y, 5))

t = np.linspace(0, 52, 300)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```
